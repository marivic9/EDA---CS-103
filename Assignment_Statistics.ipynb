# Four different types of measurement scales described in statistics: nominal, ordinal, interval, and ratio.

*Nominal*: a type of data that is used to label variables without providing any quantitative value.

*Ordinal*: a type of data that categorizes and orders variables according to their relative ranking or position.

*Interval*: a type of data that measures variables along a scale with equal intervals between successive data points.

*Ratio*: a type of data that has a true zero point and allows for the comparison of absolute quantities.

# NOMINAL

```python
import matplotlib.pyplot as plt

# Assuming 'students' is the DataFrame and 'Gender' is the column
gender_counts = students['Gender'].value_counts()

plt.figure(figsize=(8, 8))  # Set the figure size (optional)
plt.pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'pink'])
plt.title('Distribution of Students by Gender')
plt.show()
```

The pie chart depicting the gender distribution of students reveals a male-dominated population, with males representing 55% and females 45%, highlighting a significant gender gap. This imbalance raises important questions about educational equity and may influence program design to foster inclusivity while addressing biases that affect career choices in various fields. Further analysis incorporating factors such as age groups, academic programs, and socioeconomic backgrounds could provide deeper insights into the reasons behind this disparity and identify opportunities for improvement.

# ORDINAL

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='month', y='temperature', data=temp, color='skyblue')

plt.xlabel("Month")
plt.ylabel("Temperature")
plt.title("Temperature Distribution Across Months")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The temperature trend graph reveals a distinct seasonal pattern characterized by warmer temperatures from June to August and cooler conditions at the year's start and end. This variation suggests significant fluctuations between different years or locations, likely influenced by factors such as weather anomalies, climate variability, and local geography. Further analysisincorporating statistical evaluations, historical comparisons, and geographic assessmentscould enhance our understanding of these trends and elucidate underlying influences on regional climatic patterns.

# RATIO

```python
print(companies.describe())

plt.figure(figsize=(10, 6))
sns.histplot(companies['price-to-earnings ratio(TTM)'], bins=30, kde=False, color = 'skyblue')
plt.ylabel('Counts of Companies')
plt.xlabel('P/E Ratio')
plt.title('Distribution of P/E Ratios')
plt.show()

pe_counts = companies['price-to-earnings ratio(TTM)'].value_counts().sort_index()
print(pe_counts)
```

The histogram analysis of P/E ratios reveals a right-skewed distribution characterized by a significant number of companies with low or near-zero earnings, alongside outliers reflecting extremely high valuations. This suggests that while the market perceives certain firms to be strong growth candidates, many others may either be struggling financially or are overvalued relative to their earnings potential. Further investigation into descriptive statistics and industry benchmarks could provide valuable insights into underlying trends and investment opportunities within this dataset.
