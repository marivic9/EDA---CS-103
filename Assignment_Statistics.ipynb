# **INTERVAL**

```python
def potential_interval_cols(df):
  potential_interval = []
  for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
      potential_interval.append(col)
  return potential_interval

potential_interval_trainTest = potential_interval_cols(trainTest)
print("Potential interval variables in trainTest:", potential_interval_trainTest)

potential_interval_shanghaiRanking = potential_interval_cols(shanghaiRanking)
print("Potential interval variables in shanghaiRanking:", potential_interval_shanghaiRanking)

potential_interval_POP = potential_interval_cols(POP)
print("Potential interval variables in POP:", potential_interval_POP)

potential_interval_data = potential_interval_cols(data)
print("Potential interval variables in data:", potential_interval_data)

for col in potential_interval_trainTest:
  print(f"Descriptive statistics for {col} in trainTest:")
  print(trainTest[col].describe())

for col in potential_interval_shanghaiRanking:
  print(f"Descriptive statistics for {col} in shanghaiRanking:")
  print(shanghaiRanking[col].describe())

for col in potential_interval_POP:
  print(f"Descriptive statistics for {col} in POP:")
  print(POP[col].describe())

for col in potential_interval_data:
  print(f"Descriptive statistics for {col} in data:")
  print(data[col].describe())
```
