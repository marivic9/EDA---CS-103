# **EDA Visuals**

**1. Line Chart - Monthly Sales Data**

```python
msd = {
    'Month': ['January',
              'February',
              'March',
              'April',
              'May',
              'June',
              'July',
              'August',
              'September',
              'October',
              'November',
              'December'],
    'Sales (Peso)': [5000, 7000, 8000, 6000, 9000, 12000, 15000, 13000, 11000, 14000, 16000, 18000]
}

df = pd.DataFrame(msd)

plt.figure(figsize=(10, 6))
plt.plot(msd['Month'], msd['Sales (Peso)'], marker='o', linestyle='-', color='purple')
plt.grid(True)
plt.xlabel("Month")
plt.ylabel("Sales in dollars")
plt.title("Monthly Sales Data")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**2. Bar Chart - Product Sales Data**

```python
psd = {
    'Product': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E', 'Product F'],
    'Units Sold': [150, 300, 200, 400, 250, 350]
}

df = pd.DataFrame(psd)

plt.figure(figsize=(10, 6))
plt.bar(psd['Product'], psd['Units Sold'], color='purple')
plt.grid(True)
plt.xlabel("Product Names")
plt.ylabel("Number of units sold")
plt.title('Product Sales Data')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**3. Bubble Plot - Product Sales and Market Share**

```python
ps_ms = {
     "Product": ["Product A", "Product B", "Product C", "Product D", "Product E"],
     "Sales ($)": [50000, 75000, 60000, 90000, 30000],
     "Market Share (%)": [15, 25, 20, 30, 10],
     "Profit ($)": [15000, 20000, 18000, 30000, 5000]
 }
df = pd.DataFrame(ps_ms)

plt.figure(figsize=(10, 6))
plt.scatter(ps_ms["Sales ($)"], ps_ms["Market Share (%)"], s=[p / 1000 for p in ps_ms["Profit ($)"]], alpha=0.7)
plt.grid(True)
plt.xlabel('Sales ($)')
plt.ylabel('Market Share (%)')
plt.title('Product Sales and Market Share (Bubble Plot)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**4. Area plot and stacked plot - Monthly Revenue by Product Category**

```python
import matplotlib.pyplot as plt
import pandas as pd

area = {'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
        'Product A': [3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500, 7000, 7500, 8000, 8500],
        'Product B': [2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500, 7000, 7500],
        'Product C': [1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500, 7000],
        'Product D': [1000, 1200, 1500, 1800, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500]}

df = pd.DataFrame(area)

months = area['Month']
product_a = area['Product A']
product_b = area['Product B']
product_c = area['Product C']
product_d = area['Product D']

plt.figure(figsize=(15, 6))
plt.stackplot(months, product_a, product_b, product_c, product_d, labels=['Product A', 'Product B', 'Product C', 'Product D'])
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.title('Monthly Revenue by Product Category')
plt.legend(loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**5. Pie chart - Market Share by Product Category**

```python
pie = {'Product Category': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E'],
        'Market Share (%)': [25, 35, 20, 15, 5]}
df = pd.DataFrame(pie)

plt.figure(figsize=(10, 6))
plt.pie(df['Market Share (%)'], labels=df['Product Category'], autopct='%1.1f%%', startangle=90)
plt.title('Market Share by Product Category ')
plt.axis('equal')
plt.show()
```

**7. Polar Chart- Product Performance Metrics**

```python
#Data
metrics = ['Quality', 'Durability', 'Price', 'Customer Service', 'Design', 'Features']
product_a = [8, 9, 7, 6, 8, 9]
product_b = [7, 8, 6, 9, 7, 8]
product_c = [9, 7, 8, 8, 9, 6]
product_d = [6, 5, 9, 7, 8, 7]

angles = np.linspace(0, 2*np.pi, len(metrics), endpoint=False)
fig, ax = plt.subplots(figsize=(10, 11), subplot_kw=dict(polar=True))

ax.plot(angles, product_a, 'r-', label="Product A")
ax.plot(angles, product_b, 'g-', label="Product B")
ax.plot(angles, product_c, 'b-', label="Product C")
ax.plot(angles, product_d, 'm-', label="Product D")

ax.set_thetagrids(angles * 180/np.pi, metrics)
ax.set_title("Product Performance Metrics (Polar Chart)")

plt.legend(loc='upper right')
plt.show()
```

**8. Lollipop Chart- Product Sales Figures**

```python
#DataFrame
data = {
    'Product': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E', 'Product F'],
    'Sales (Peso)': [5000, 8000, 6500, 12000, 7000, 9000]
}
df = pd.DataFrame(data)


df = df.sort_values('Sales (Peso)', ascending=True)
fig, ax = plt.subplots(figsize=(12, 6))

ax.vlines(x=df.index, ymin=0, ymax=df['Sales (Peso)'], color='firebrick', alpha=0.7, linewidth=2)
ax.scatter(x=df.index, y=df['Sales (Peso)'], s=75, color='firebrick', alpha=0.7)

ax.set_title('Product Sales Figures', fontdict={'size': 22})
ax.set_ylabel('Sales (Peso)')
ax.set_xlabel('Product Names')
ax.set_xticks(df.index)
ax.set_xticklabels(df['Product'], rotation=45, ha='right')
ax.set_ylim(0, max(df['Sales (Peso)']) * 1.1)

for i, v in enumerate(df['Sales (Peso)']):
    ax.text(i, v + 200, str(v), ha='center', va='bottom')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
